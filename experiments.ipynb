```python
def method_decorator(func):
    def say_hello1():
        print("Before method execution")
        res = func()
        print("After method execution")
        return res

    return say_hello1
    


        
@method_decorator
def say_hello():
        return "Hello!"

say_hello()
```

```python
def process_items(items: list[str]):
    for item in items:
        print(type(item))

    print('----------------')
    print(type(items))



process_items('"apple", "banana", "cherry"')
```
